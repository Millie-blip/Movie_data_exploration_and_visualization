```python
# Load the CSV file
data = pd.read_csv(r"C:/Users/user/Downloads/archive (2)/tmdb_5000_movies.csv")

# Function to parse JSON and combine 'name' values
def parse_and_combine(json_str):
    try:
        # Parse the JSON string
        data = json.loads(json_str)
        # Extract and combine the 'name' values
        names = [item['name'] for item in data]
        return ', '.join(names)
    except (json.JSONDecodeError, TypeError):
        return None

# List of columns containing JSON strings to be processed
json_columns = ['genres', 'keywords', 'production_companies', 'production_countries', 'spoken_languages']

# Apply the parsing function to each specified column
for column in json_columns:
    data[column] = data[column].apply(parse_and_combine)

# Save the DataFrame back to a CSV file
output_file_path = 'output_file.csv'  # Replace with your desired output file path
data.to_csv(output_file_path, index=False)

# Print the resulting DataFrame to verify
print(data[json_columns])
```

```python
def process_genres(x):
    # Check if x is a string
    if isinstance(x, str):
        return x.split(', ')
    # Check if x is already a list or other iterable
    elif isinstance(x, (list, tuple)):
        return list(x)
    # Handle missing or non-standard cases
    elif pd.isna(x):
        return []
    else:
        # For unexpected types, convert to string if possible or handle as empty
        try:
            return str(x).split(', ')
        except Exception:
            return []

# Apply the function to the genres and keywords columns
data['genres'] = data['genres'].apply(process_genres)
data['keywords'] = data['keywords'].apply(process_genres)
data['production_companies'] = data['production_companies'].apply(process_genres)
data['production_countries'] = data['production_countries'].apply(process_genres)
data['spoken_languages'] = data['spoken_languages'].apply(process_genres)

# Check the results
print(data['genres'].head())
print(data['keywords'].head())
print(data['production_companies'].head())
print(data['production_countries'].head())
print(data['spoken_languages'].head())
```

```python
data['overview'].fillna('No overview available', inplace=True)
data['tagline'].fillna('No tagline', inplace=True)
```

```python
data['homepage'].fillna('No homepage', inplace=True)
```

```python
data.dropna(subset=['release_date'], inplace=True)
```

```python
data['runtime'].fillna(data['runtime'].median(), inplace=True)
```

```python
print(data.isnull().sum())
```

```python
# Converting release date to datetime
data['release_date'] = pd.to_datetime(data['release_date'])

# Extracting year from release date
data['release_year'] = data['release_date'].dt.year
```

```python
# Identify columns that have list-like data
list_columns = [col for col in data.columns if data[col].apply(lambda x: isinstance(x, list)).any()]
print(f"Columns containing lists: {list_columns}")

# Convert all list-like columns to strings
for col in list_columns:
    data[col] = data[col].apply(str)

# Now check for duplicates
duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
import ast

# Convert string representations of lists into actual lists (if needed)
for col in list_columns:
    data[col] = data[col].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else x)

# Function to convert lists to comma-separated strings
def convert_list_to_string(x):
    if isinstance(x, list):
        return ', '.join(map(str, x))  # Convert each element to a string and join with a comma
    return x

# Now apply the list-to-string conversion
for col in list_columns:
    data[col] = data[col].apply(convert_list_to_string)

# Save the updated DataFrame back to CSV
data.to_csv('clean_movie_dataset.csv', index=False)

# Display the first few rows of the modified DataFrame
data.head()


```
