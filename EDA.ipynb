```python
#Import python libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('clean_movie_dataset.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# 1. Genre Trends
data['genres'] = data['genres'].apply(lambda x: x.split(', ') if pd.notnull(x) else [])
genre_exploded = data.explode('genres')
genre_counts = genre_exploded['genres'].value_counts()
plt.figure(figsize=(10,6))
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.title('Most Common Movie Genres')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Rating Distribution
plt.figure(figsize=(10,6))
sns.histplot(data['vote_average'], bins=10, kde=True)
plt.title('Distribution of Movie Ratings')
plt.show()
```

```python
# 4. Release Year Trends
plt.figure(figsize=(10,6))
sns.histplot(data['release_year'], bins=50, kde=False)
plt.title('Number of Movies Released Per Year')
plt.show()
```

```python
# 5. Box Office Trends
plt.figure(figsize=(10,6))
sns.scatterplot(x='release_year', y='revenue', data=data)
plt.title('Box Office Revenue Over Time')
plt.show()
```
